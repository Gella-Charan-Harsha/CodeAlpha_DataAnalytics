```python
import pandas as pd

df = pd.read_csv('../datasets/books.csv')

print(df.head())
print('--- df.info() ---')
df.info()
print('--- df.describe() ---')
print(df.describe())
print('--- df.shape ---')
print(df.shape)
```

```python
# Answer questions requested
total_books = len(df)
avg_price = df['Price'].astype(float).mean()
highest_price = df['Price'].astype(float).max()
lowest_price = df['Price'].astype(float).min()
rating_distribution = df['Rating'].value_counts()
missing_values = df.isnull().sum()

print('Total Books:', total_books)
print('Average Price:', avg_price)
print('Highest Price:', highest_price)
print('Lowest Price:', lowest_price)
print('\nRating Distribution:\n', rating_distribution)
print('\nMissing Values:\n', missing_values)
```
