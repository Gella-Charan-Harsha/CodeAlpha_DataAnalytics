```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download vader lexicon (only needed once)
nltk.download('vader_lexicon')

# Replace this with your Kaggle Amazon Reviews CSV path
# Example: reviews = pd.read_csv('../datasets/amazon_reviews.csv')
reviews = None
text_col = None  # set to the column containing the review text (e.g., 'reviewText' or 'review')

print('Load your reviews CSV into `reviews` and set `text_col`.')

sia = SentimentIntensityAnalyzer()

def classify_sentiment(text: str) -> str:
    if text is None:
        return 'Neutral'
    scores = sia.polarity_scores(str(text))
    compound = scores.get('compound', 0)
    if compound >= 0.05:
        return 'Positive'
    if compound <= -0.05:
        return 'Negative'
    return 'Neutral'

# Uncomment after you load data
# reviews['sentiment'] = reviews[text_col].apply(classify_sentiment)
#
# sentiment_counts = reviews['sentiment'].value_counts(normalize=True) * 100

print('After loading, run the classification line and the plotting section below.')
```
